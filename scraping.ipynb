```python
# def scrape(Category_name, url):
#     driver.get(url)
#     time.sleep(2)
    
#     # txt = driver.find_elements_by_tag_name('body').text
#     # img = [img.get_attribute('src') for img in driver.find_elements_by_tag_name('img')]
#     # links = [link.get_attribute("href") for link in driver.find_elements_by_tag_name('a')]
#     txt = driver.find_element(By.TAG_NAME, 'p')
#     txt = txt.text
#     img = driver.find_element(By.XPATH, 'img')
#     img = img.get_attribute('src')
#     links = driver.find_element(By.TAG_NAME, 'a')
#     links = links.get_attribute('href')
#     # paragraph = WebDriverWait(driver, 10).until(
#     #     EC.visibility_of_element_located((By.TAG_NAME, 'p'))
#     # )
#     # txt = paragraph.text
    
#     # # Wait for the image element to be visible
#     # image = WebDriverWait(driver, 10).until(
#     #     EC.visibility_of_element_located((By.TAG_NAME, 'img'))
#     # )
#     # img_src = image.get_attribute('src')
    
#     # # Wait for the link element to be visible
#     # link = WebDriverWait(driver, 10).until(
#     #     EC.visibility_of_element_located((By.TAG_NAME, 'a'))
#     # )
#     # link_href = link.get_attribute('href')

#     return {
#         "Category" : Category_name,
#         "url" : url,
#         "Text" : txt,
#         "Images" : img,
#         "Links" : links
#     }

# def scrape_websites(urls):
#   # Set Chrome options (optional: headless mode)
#   options = webdriver.ChromeOptions()
#   # options.add_argument('--headless')  # Uncomment for headless mode

#   # Initialize Chrome webdriver
#   driver = webdriver.Chrome(options=options)
#   scraped_data = []

#   try:
#     for url in urls:
#       # Open the website
#       driver.get(url)

#       # Extract text (modify selector for specific needs)
#       text_elements = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.TAG_NAME, "p")))  # Wait for element
#       text = text_elements.text

#       # Extract links
#       links = [element.get_attribute("href") for element in driver.find_elements(By.TAG_NAME, "a")]

#       # Extract images (source URLs)
#       images = [element.get_attribute("src") for element in driver.find_elements(By.TAG_NAME, "img")]

#       # Append data for this website
#       scraped_data.append({
#           "url": url,
#           "text": text,
#           "links": links,
#           "images": images
#       })

#   except Exception as e:
#     print(f"Error scraping website: {url}")
#     print(e)
#     pass

#   finally:
#     # Close the browser window
#     driver.quit()

#   return scraped_data
def scrape_websites(urls):
    # Set Chrome options (optional: headless mode)
    options = webdriver.ChromeOptions()
    # options.add_argument('--headless')  # Uncomment for headless mode

    # Initialize Chrome webdriver
    driver = webdriver.Chrome(options=options)
    scraped_data = []

    try:
        for url in urls:
            try:
                # Open the website
                driver.get(url)

                # Extract text (modify selector for specific needs)
                text_elements = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.TAG_NAME, "p")))  # Wait for element
                text = text_elements.text

                # Extract links
                links = [element.get_attribute("href") for element in driver.find_elements(By.TAG_NAME, "a")]

                # Extract images (source URLs)
                images = [element.get_attribute("src") for element in driver.find_elements(By.TAG_NAME, "img")]

                # Append data for this website
                scraped_data.append({
                    "url": url,
                    "text": text,
                    "links": links,
                    "images": images
                })
            except Exception as e:
                print(f"Error scraping website: {url}")
                print(e)
                pass

    finally:
        # Close the browser window
        driver.quit()

    return scraped_data
```
